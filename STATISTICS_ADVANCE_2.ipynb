#Question 1: What is hypothesis testing in statistics?

-    Answer: Hypothesis testing is a formal statistical framework used to make inferences or decisions about a population parameter based on sample data. It allows researchers to determine whether there is enough statistical evidence to support a specific claim or theory.The process involves formulating two competing statements (the null and alternative hypotheses), collecting sample data, calculating a test statistic, and using a decision rule to either reject or fail to reject the default assumption. It is widely used in science, engineering, business, and medicine to validate experimental results and ensure that observed effects are not just the result of random chance.

#Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?   

-  Answer:The Null Hypothesis ($H_0$) and the Alternative Hypothesis ($H_1$ or $H_a$) are mutually exclusive statements that form the foundation of a statistical test.Key Differences:FeatureNull Hypothesis (H0​)Alternative Hypothesis (H1​)Core MeaningRepresents the status quo, no effect, no difference, or no change.Represents the claim, effect, difference, or relationship you want to prove.Mathematical SymbolsTypically contains equality signs ($=, \le, \ge$).Typically contains inequality signs ($\ne, <, >$).Burden of ProofAssumed to be true by default until proven otherwise.Bears the burden of proof; only accepted if $H_0$ is rejected.Example: If you are testing a new medical drug, $H_0$ would state that the drug has no effect on patients compared to a placebo. $H_1$ would state that the drug is effective and produces a significantly different health outcome.

#Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.   

-  Answer:The significance level, denoted by the Greek letter alpha ($\alpha$), is the threshold set by the researcher before conducting a test to define what constitutes a "statistically significant" result. It represents the probability of rejecting the null hypothesis when it is actually true (a Type I error). Common choices for $\alpha$ are $0.05$ (5%), $0.01$ (1%), or $0.10$ (10%).Role in Deciding the Outcome:During a hypothesis test, you calculate a p-value, which measures the probability of obtaining sample results at least as extreme as the ones observed, assuming the null hypothesis is true.The decision rule is straightforward:If $\text{p-value} \le \alpha$: The result is statistically significant. You reject $H_0$ and support $H_1$.If $\text{p-value} > \alpha$: The result is not statistically significant. You fail to reject $H_0$.For instance, if $\alpha = 0.05$ and your test yields a p-value of $0.02$, the probability of getting that data by pure chance is only 2%. Since $0.02 \le 0.05$, you reject the null hypothesis.

#Question 4: What are Type I and Type II errors? Give examples of each.

-Answer:No statistical test is 100% certain because they rely on sample data. Consequently, two types of errors can occur:Type I Error (False Positive)A Type I error occurs when you reject the null hypothesis when it is actually true. The probability of committing this error is equal to the significance level ($\alpha$).Example: A medical test falsely indicates that a healthy person has a disease. (The null hypothesis of "no disease" is rejected incorrectly).Type II Error (False Negative)A Type II error occurs when you fail to reject the null hypothesis when it is actually false. The probability of committing this error is denoted by beta ($\beta$).Example: A medical test fails to detect a disease in a person who actually has it. (The null hypothesis of "no disease" is accepted incorrectly).

#Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.


-Answer:Both Z-tests and T-tests are parametric statistical tests used to compare sample means, but they are applied under different conditions regarding sample size and population parameter knowledge.Key Differences & When to Use:Z-test: Used when the population standard deviation ($\sigma$) is known and/or the sample size is large ($n \ge 30$). It relies on the standard normal distribution ($Z$-distribution).T-test: Used when the population standard deviation ($\sigma$) is unknown (so you use the sample standard deviation $s$ instead) and the sample size is small ($n < 30$). It relies on Student's $t$-distribution, which has heavier tails to account for extra uncertainty.Decision Matrix:Population variance ($\sigma^2$) known + Large or small sample $\rightarrow$ Z-testPopulation variance ($\sigma^2$) unknown + Large sample ($n \ge 30$) $\rightarrow$ Z-test (via Central Limit Theorem) or T-test (more conservative)Population variance ($\sigma^2$) unknown + Small sample ($n < 30$) $\rightarrow$ T-test

```python
#QUESTION6-Question 6: Write a Python program to generate a binomial distribution with $n=10$ and $p=0.5,$ then plot its histogram.
'''ANSWER'''
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters for the binomial distribution
n = 10   # number of trials
p = 0.5  # probability of success
size = 10000  # number of random samples to generate

# Generate random numbers simulating a binomial distribution
binomial_data = np.random.binomial(n, p, size)

# Plotting the histogram
plt.figure(figsize=(8, 5))
# Aligned bins to center the integers 0 through 10 nicely
plt.hist(binomial_data, bins=np.arange(-0.5, 11.5, 1), rwidth=0.8, color='skyblue', edgecolor='black', density=True)

plt.title(f'Binomial Distribution Histogram (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability Density')
plt.xticks(range(0, 11))
plt.grid(axis='y', alpha=0.75)

# Show plot
plt.show()
```

```python
#Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python
'''ANSWER'''
import numpy as np
from statsmodels.stats.weightstats import ztest as ztest

# Given sample dataset
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean (Let's test if the true population mean is 50.0)
hypothesized_mean = 50.0

# Performing a two-sided one-sample Z-test
z_stat, p_value = ztest(sample_data, value=hypothesized_mean)

print(f"Sample Mean: {np.mean(sample_data):.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value:     {p_value:.4f}")

# Interpreting the results at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print(f"\nResult: Reject the null hypothesis. The sample mean is significantly different from {hypothesized_mean}.")
else:
    print(f"\nResult: Fail to reject the null hypothesis. There is no significant difference from {hypothesized_mean}.")
```

```python
#Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Simulate data from a Normal Distribution (mean=100, std_dev=15, size=100)
true_mean = 100
true_std = 15
sample_size = 100
data = np.random.normal(true_mean, true_std, sample_size)

# Calculate sample statistics
sample_mean = np.mean(data)
sample_sem = stats.sem(data) # Standard Error of the Mean

# Calculate 95% Confidence Interval for the mean using t-distribution
confidence_level = 0.95
ci_lower, ci_upper = stats.t.interval(confidence_level, df=sample_size-1, loc=sample_mean, scale=sample_sem)

print(f"Sample Mean: {sample_mean:.4f}")
print(f"95% Confidence Interval for the Mean: ({ci_lower:.4f}, {ci_upper:.4f})")

# Plotting the data
plt.figure(figsize=(9, 5))
plt.hist(data, bins=15, color='lightgreen', edgecolor='black', alpha=0.7, label='Simulated Data')
plt.axvline(sample_mean, color='blue', linestyle='dashed', linewidth=2, label=f'Sample Mean ({sample_mean:.2f})')
plt.axvline(ci_lower, color='red', linestyle='dotted', linewidth=2, label=f'95% CI Lower ({ci_lower:.2f})')
plt.axvline(ci_upper, color='red', linestyle='dotted', linewidth=2, label=f'95% CI Upper ({ci_upper:.2f})')

plt.title('Simulation of Normal Data & 95% Confidence Interval for Mean')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
#QUESTION9 Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores representin terms of standard deviations from the mea
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(dataset):
    # Convert input list to numpy array if necessary
    dataset = np.array(dataset)

    # Calculate Mean and Standard Deviation
    mean = np.mean(dataset)
    std_dev = np.std(dataset)

    # Calculate Z-scores
    z_scores = (dataset - mean) / std_dev

    # Visualizing the standardized data
    plt.figure(figsize=(8, 5))
    plt.hist(z_scores, bins=10, color='salmon', edgecolor='black', alpha=0.8)

    plt.title('Histogram of Standardized Data (Z-scores)')
    plt.xlabel('Z-score (Standard Deviations from Mean)')
    plt.ylabel('Frequency')
    plt.axvline(0, color='black', linestyle='dashed', linewidth=1.5, label='Mean (Z=0)')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    return z_scores

# Synthetic dataset example to run the function
example_data = [55, 67, 72, 81, 93, 45, 77, 83, 89, 64, 71, 75, 88, 92, 60]
z_results = calculate_and_plot_zscores(example_data)
#EXPLAINATION:A Z-score of 0 implies the data point is exactly equal to the mean.A positive Z-score (e.g., $+1.5$) indicates the data point lies $1.5$ standard deviations above the mean.A negative Z-score (e.g., $-2.0$) indicates the data point lies $2.0$ standard deviations below the mean
```
